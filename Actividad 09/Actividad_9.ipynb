**Calculamos MZsR a z=0 marcando las galaxias seleccionadas:**

```python
table_a = pd.read_csv(path_lin+'/Aperturas_RefL0100N1504_snap_28_aperture_30.dat')
Ref     = table_a[(table_a['Stars_Mass']>=1e10)].copy()

SF_O  = Ref['SF_Oxygen']
SF_H  = Ref['SF_Hydrogen']
SF_M  = Ref['SF_Mass']
M_Star= Ref['Stars_Mass']

M_O_SF = SF_O*SF_M
M_H_SF = SF_H*SF_M

Ref['sf_metalicity'] = 12 + np.log10(M_O_SF/(16*M_H_SF))
```

**Calculamos la media:**

```python
mask = (Ref['sf_metalicity'].notna())
xvar = np.log10(Ref['Stars_Mass'])[mask].to_numpy()
yvar = Ref['sf_metalicity'][mask].to_numpy()
```

```python
nbins=10        # Numero de bins en variable x a usar para dividir la muestra
perc_low=25.        # Percentil a usar como cota inferior de error
perc_high=75.       # Percentil a usar como cota superior de error
min_count=10         # Mínima cantidad de elementos para considerar en un bin

xmedian,ymedian,xerror,yerror,elements=Curves2D.median_curve(xvar,yvar,bins=nbins,min_count=min_count,
                                                             perc_low=perc_low,perc_high=perc_high)
```

```python
mask_1 = (Ref['sf_metalicity'].notna())&(Ref['GalaxyID'] == 14696151)
mask_2 = (Ref['sf_metalicity'].notna())&(Ref['GalaxyID'] == 14875627)

xvar_1 = np.log10(Ref['Stars_Mass'])[mask_1].to_numpy()
yvar_1 = Ref['sf_metalicity'][mask_1].to_numpy()

xvar_2 = np.log10(Ref['Stars_Mass'])[mask_2].to_numpy()
yvar_2 = Ref['sf_metalicity'][mask_2].to_numpy()

mask_3 = (Ref['sf_metalicity'].notna())&(Ref['GalaxyID'] == 16062294)
mask_4 = (Ref['sf_metalicity'].notna())&(Ref['GalaxyID'] == 15457919)

xvar_3 = np.log10(Ref['Stars_Mass'])[mask_3].to_numpy()
yvar_3 = Ref['sf_metalicity'][mask_3].to_numpy()

xvar_4 = np.log10(Ref['Stars_Mass'])[mask_4].to_numpy()
yvar_4 = Ref['sf_metalicity'][mask_4].to_numpy()
```

```python
fig,ax=plt.subplots(figsize=(8,6))

xlim=(min(xvar),max(xvar)+0.2)
ylim=(min(yvar)+0.2,max(yvar)+0.2)

xlabel=r'$Log(M_{Star})[M_{Sun}]$'
ylabel=r'$12 + Log(O/H)$'

ax.set_xlabel(xlabel,fontsize=18)
ax.set_ylabel(ylabel,fontsize=18)

ax.set_xlim(xlim)
ax.set_ylim(ylim)

ax.tick_params(axis='both', which='major', labelsize=15)
                
plt.subplots_adjust(left=0.12,right=0.95,bottom=0.15,top=0.95)

# scatter plot
ax.scatter(xvar,yvar,marker='o',s=0.9,c='0.3',label='Scatter plot')

ax.scatter([xvar_1,xvar_2],[yvar_1,yvar_2],marker='o',s=20,c='b')

ax.scatter([xvar_3,xvar_4],[yvar_3,yvar_4],marker='o',s=20,c='r')


ax.plot(xmedian,ymedian,c='k',marker='o',ms=8,linestyle='-',linewidth=1.5,label='Mediana ')

ax.errorbar(xmedian,ymedian,yerr=yerror,
           c='k',marker='o',markersize=8,linestyle='-',
           linewidth=1.5,label='Percentiles 25º y 75º',capsize=5,capthick=1.5)

leg1=ax.legend(edgecolor='0.1',loc='lower right',handlelength=4,fontsize=12)
leg1.set_title(r'$\rm{EAGLE\ RefL0100N1504},\ z=0$',prop={'size':15})

plt.grid()
# ax.add_artist(leg1)
# plt.savefig(path_lin_fig+'scatter_log(M)vs_metalicidad.png')


plt.show()
```
