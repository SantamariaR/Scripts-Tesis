**Importamos algunas librerías** 

```python
import matplotlib.pyplot as plt
import eagleSqlTools as sql
import numpy as np
from astropy.io import ascii
from astropy.table import Table
```

Éste código va a ser para el barrido en redshift

```python
simu='RefL0100N1504'
```

```python
data_folder='/home/ramiro/Facultad/Tesis/Datos/Actividad 2/'
fig,ax=plt.subplots(figsize=(8,6))
snap_list=[28,19,15,12]
redshift_list=[0,1,2,3]
simu='RefL0100N1504'

color=['k','b','g','r']    
line=['-','--','-.',':']   

for snap,redshift,lc,ls in zip(snap_list,redshift_list,color,line):   
    data_file='evol_con_z_'+simu+'_snap_'+str(snap)+'.dat'
    data_path=data_folder+data_file
    table=ascii.read(data_path,comment='#',format='fast_csv') 
         
    mask_metallicity = (table['SF_Metallicity']>0)
    SF_M   = table['SF_Metallicity'][mask_metallicity]
    Z      = 0.02
    
    var   = np.log10(SF_M/Z) 
    
    weights=np.ones_like(var)/float(len(var))

    ax.hist(var,bins=15,range=(-1,1),histtype='step',label='z='+str(redshift),
            linestyle=ls,color=lc,linewidth=2,weights=weights)

ax.set_xlabel(r'$Log(SF_{Met}/Z_{Sun})$', fontsize=15)
ax.set_ylabel('Fraction', fontsize=15)
ax.tick_params(axis='both',labelsize=15)
plt.title('SF gas-phase metallicity mayores a 'r'$10^{10}$ masas solares')

leg=plt.legend(loc='upper left',edgecolor='w',fontsize=11)
leg.set_title('EAGLE RefL0100N1504', prop = {'size':13})
plt.grid(True)
#plt.savefig('Log(SF_Met-Z_Sun)_(z=0,1,2,3).png')
```

**Ahora pasamos a graficar diferentes catidades al mismo Redshift(z=0) pero variando el aperture size(5, 10, 30, 70 y 100 pkpc)**

```python
data_folder='/home/ramiro/Facultad/Tesis/Datos/Actividad 2/'
fig,ax=plt.subplots(figsize=(8,6))
simu='RefL0100N1504'

aperture_list=[5,10,30,70,100]

color=['k','b','g','r','c']    
line=['-','--','-.',':','-.']   

for aper,lc,ls in zip(aperture_list,color,line):   
    
    data_file='Aperturas_'+simu+'_snap_28_aperture_'+str(aper)+'.dat'
    data_path=data_folder+data_file
    table=ascii.read(data_path,comment='#',format='fast_csv') 
    
    M_Star = table['Mass_Star']
    
    var    =  np.log10(M_Star)
    
    weights=np.ones_like(var)/float(len(var))

    ax.hist(var,bins=15,range=(9,12.5),histtype='step',label='Aperture='+str(aper),
            linestyle=ls,color=lc,linewidth=2,weights=weights)

ax.set_xlabel(r'$Log(M_\bigstar) \ [\rm{M}_\odot]$', fontsize=15)
ax.set_ylabel('Fraction', fontsize=15)
ax.tick_params(axis='both',labelsize=15)
plt.title('Log(M_Star) para gal. mayores a 'r'$10^{10}$ masas solares en diferentes aperturas[pkpc]')

leg=plt.legend(loc='upper right',edgecolor='w',fontsize=13)
leg.set_title('EAGLE RefL0100N1504', prop = {'size':15})
plt.grid(True)
#plt.savefig('Log(M_Stars)-(Apertures=5,10,30,70,100).png')
```

```python
data_folder='/home/ramiro/Facultad/Tesis/Datos/Actividad 2/'
fig,ax=plt.subplots(figsize=(8,6))
simu='RefL0100N1504'

aperture_list=[5,10,30,70,100]

color=['k','b','g','r','c']    
line=['-','--','-.',':','-.']   

for aper,lc,ls in zip(aperture_list,color,line):   
    
    data_file='Aperturas_'+simu+'_snap_28_aperture_'+str(aper)+'.dat'
    data_path=data_folder+data_file
    table=ascii.read(data_path,comment='#',format='fast_csv') 
    
    mask_SFR=(table['SFR']>0.0)
    SFR   = table['SFR'][mask_SFR]
    M_Star= table['Mass_Star'][mask_SFR]
    sSFR  = SFR/M_Star
    
    var    =  np.log10(sSFR)
    
    weights=np.ones_like(var)/float(len(var))

    ax.hist(var,bins=15,range=(-14,-9),histtype='step',label='Aperture='+str(aper),
            linestyle=ls,color=lc,linewidth=2,weights=weights)

ax.set_xlabel(r'$Log(sSFR)$', fontsize=15)
ax.set_ylabel('Fraction', fontsize=15)
ax.tick_params(axis='both',labelsize=15)
plt.title('Log(sSFR) para gal. mayores a 'r'$10^{10}$ masas solares en diferentes aperturas[pkpc]')

leg=plt.legend(loc='upper left',edgecolor='w',fontsize=13)
leg.set_title('EAGLE RefL0100N1504', prop = {'size':15})
plt.grid(True)
#plt.savefig('Log(sSFR)-(Apertures=5,10,30,70,100).png')
```
