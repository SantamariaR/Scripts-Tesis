## Histogramas de Level of merger

### Level of Mergers entre 1 y 2

```python
width=0.05   #Ancho de los bins

figure, axes = plt.subplots(14,2,figsize=(20,90))

# Convertimos a ax en una array plano para recorrerlo tranqui con un for
axes = axes.flatten() 

figure.tight_layout(pad=12.0)

fontdict ={'fontsize': 30}

# figure.suptitle(r'$Histogramas\ de\ Level\ of\ Merger\ entre\ 1\ y\ 2\ para\ diferentes\ redshift.$', fontsize=30)

bins = np.arange(1,2,width)

for i in np.arange(len(merger.columns[2:])):
    mask_Lm = (merger[merger.columns[2:][i]]>1)&(merger[merger.columns[2:][i]]<=2)
    var  = merger[merger.columns[2:][i]][mask_Lm].values

    h_var,b = np.histogram(var,bins,density=True)
    axes[i].bar(b[:-1],height=h_var * np.diff(b),width =width,color='pink',edgecolor='purple',capsize=4,label='Cantidad\ de\ galaxias:'+str(len(var)))
    axes[i].set_title(r'$Histograma\ en\ z='+REDSHIFT[i]+'.$',fontsize=20)
    axes[i].set_xlabel(r'$L_{merger}$',fontsize=20)
    axes[i].set_ylabel(r'$Probabilidad$',fontsize=20)
    axes[i].set_xticks(bins)
    
    leg=axes[i].legend(loc='upper right',fontsize=15,handlelength=5)

    axes[i].set_xticklabels(np.round(bins,2),fontsize=10,rotation=60)

# figure.savefig('1<Mergers=<2.png')
plt.show()
```

### Level of merger mayores a 2

```python
width=10   #Ancho de los bins

figure, axes = plt.subplots(14,2,figsize=(20,90))

# Convertimos a ax en una array plano para recorrerlo tranqui con un for
axes = axes.flatten() 

figure.tight_layout(pad=9.0)

fontdict ={'fontsize': 30}

# figure.suptitle(r'$Histogramas\ de\ Level\ of\ Merger\ entre\ 1\ y\ 2\ para\ diferentes\ redshift.$', fontsize=30)
bins = np.arange(2,150,width)

for i in np.arange(len(merger.columns[2:])):
    mask_Lm = (merger[merger.columns[2:][i]]>2)
    var = merger[merger.columns[2:][i]][mask_Lm].values

    h_var,b = np.histogram(var,bins,density=True)
    axes[i].bar(b[:-1],height=h_var * np.diff(b),width =width,color='pink',edgecolor='purple',capsize=4,label='Cantidad\ de\ galaxias:'+str(len(var)))
    axes[i].set_title(r'$Histograma\ en\ z='+REDSHIFT[i]+'.$',fontsize=20)
    axes[i].set_xlabel(r'$L_{merger}$',fontsize=20)
    axes[i].set_ylabel(r'$Probabilidad$',fontsize=20)
    axes[i].set_xticks(bins)
    
    leg=axes[i].legend(loc='upper right',fontsize=15,handlelength=5)

    axes[i].set_xticklabels(np.round(bins,2),fontsize=10,rotation=60)
    

figure.savefig('Mergers>2')  
plt.show()
```
