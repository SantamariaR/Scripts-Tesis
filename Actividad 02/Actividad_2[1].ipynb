## Log(M_Star)

```python
simu='RefL0100N1504'
```

```python
data_folder='/home/ramiro/Facultad/Tesis/Datos/Datos_tesis/'
fig,ax=plt.subplots(figsize=(10,8))
snap_list=[28,19,15,12]
redshift_list=[0,1,2,3]
simu='RefL0100N1504'

color=['k','b','g','r']    
line=['-','--','-.',':']   

for snap,redshift,lc,ls in zip(snap_list,redshift_list,color,line):   
    data_file='datos_tesis_'+simu+'_snap_'+str(snap)+'_aprture_ninguna.dat'
    data_path=data_folder+data_file
    table=ascii.read(data_path,comment='#',format='fast_csv') 
    
    Ms = table['Stars_Mass']
    
    var   = np.log10(Ms) 
    
    weights=np.ones_like(var)/float(len(var))

    ax.hist(var,bins=15,range=(10,11.8),histtype='step',label='z='+str(redshift),linestyle=ls,color=lc,linewidth=3,weights=weights)

ax.set_xlabel(r'$log(M_\bigstar) \ [\rm{M}_\odot]$', fontsize=30)
ax.set_ylabel(r'$Fracción$', fontsize=30)
ax.tick_params(axis='both',labelsize=25)
# plt.title('SF gas-phase metallicity mayores a 'r'$10^{10}$ masas solares')

leg=plt.legend(loc='upper right',edgecolor='w',fontsize=25)
# leg.set_title('EAGLE RefL0100N1504', prop = {'size':20})
plt.grid(True)
plt.subplots_adjust(left=0.3, right=0.9, top=0.9, bottom=0.1)

# plt.savefig('histo_log(M_star)_(z=0,1,2,3).pdf')
```

## $12+log(O/H)$

```python
data_folder='/home/ramiro/Facultad/Tesis/Datos/Datos_tesis/'
fig,ax=plt.subplots(figsize=(10,8))
snap_list=[28,19,15,12]
redshift_list=[0,1,2,3]
simu='RefL0100N1504'

color=['k','b','g','r']    
line=['-','--','-.',':']   

for snap,redshift,lc,ls in zip(snap_list,redshift_list,color,line):   
    data_file='datos_tesis_'+simu+'_snap_'+str(snap)+'_aprture_ninguna.dat'
    data_path=data_folder+data_file
    table=ascii.read(data_path,comment='#',format='fast_csv') 
     
    mask = (table['SF_Oxygen']>0)&(table['SF_Hydrogen']>0) 
    var  = 12 + np.log10(table['SF_Oxygen'][mask])-np.log10(table['SF_Hydrogen'][mask])-np.log10(16)
    
    weights=np.ones_like(var)/float(len(var))

    ax.hist(var,bins=15,range=(8.1,9.75),histtype='step',label='z='+str(redshift),
            linestyle=ls,color=lc,linewidth=3,weights=weights)

ax.set_xlabel(r'$12 + log(O/H)$', fontsize=30)
ax.set_ylabel(r'$Fracción$', fontsize=30)
ax.tick_params(axis='both',labelsize=25)
# plt.title('SF gas-phase metallicity mayores a 'r'$10^{10}$ masas solares')

leg=plt.legend(loc='upper right',edgecolor='w',fontsize=25)
# leg.set_title('EAGLE RefL0100N1504', prop = {'size':13})
plt.grid(True)

plt.subplots_adjust(left=0.3, right=0.9, top=0.9, bottom=0.1)

# plt.savefig('histo_metalicidad_(z=0,1,2,3).pdf')
```

```python
data_folder='/home/ramiro/Facultad/Tesis/Datos/Actividad 2/'
fig,ax=plt.subplots(figsize=(8,6))
simu='RefL0100N1504'

aperture_list=[5,10,30,70,100]

color=['k','b','g','r','c']    
line=['-','--','-.',':','-.']   

for aper,lc,ls in zip(aperture_list,color,line):   
    
    data_file='Aperturas_'+simu+'_snap_28_aperture_'+str(aper)+'.dat'
    data_path=data_folder+data_file
    table=ascii.read(data_path,comment='#',format='fast_csv') 
    
    mask_SFR=(table['SFR']>0.0)
    SFR   = table['SFR'][mask_SFR]
    M_Star= table['Mass_Star'][mask_SFR]
    sSFR  = SFR/M_Star
    
    var    =  np.log10(sSFR)
    
    weights=np.ones_like(var)/float(len(var))

    ax.hist(var,bins=15,range=(-14,-9),histtype='step',label='Aperture='+str(aper),
            linestyle=ls,color=lc,linewidth=2,weights=weights)

ax.set_xlabel(r'$Log(sSFR)$', fontsize=15)
ax.set_ylabel('Fraction', fontsize=15)
ax.tick_params(axis='both',labelsize=15)
plt.title('Log(sSFR) para gal. mayores a 'r'$10^{10}$ masas solares en diferentes aperturas[pkpc]')

leg=plt.legend(loc='upper left',edgecolor='w',fontsize=13)
leg.set_title('EAGLE RefL0100N1504', prop = {'size':15})
plt.grid(True)
#plt.savefig('Log(sSFR)-(Apertures=5,10,30,70,100).png')
```
