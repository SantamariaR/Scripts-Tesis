```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
```

```python
DF = pd.read_csv('galaxy_massive_project_RefL0100N1504_snap_28.dat')
```

```python
DF_NEW = DF.copy()
```

```python
x = DF_NEW.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
DF_NEW_1 = pd.DataFrame(x_scaled,columns =['Stars_Mass', 'SF_Hydrogen','SF_Oxygen', 'SF_Mass', 'NSF_Mass', 'BlackHoleMass','StarFormationRate', 'KappaCoRot'])
```

```python
DF_NEW.hist(edgecolor='black',bins = 20 ,linewidth=0.2)
fig=plt.gcf()
fig.set_size_inches(15,15)
plt.show()
```

```python
X = DF_NEW.iloc[:,:]
f, ax = plt.subplots(figsize=(10, 8))
corr_1 = X.corr()
print(corr_1)
sns.heatmap(corr_1, mask=np.zeros_like(corr_1), 
          cmap=sns.diverging_palette(220, 10, as_cmap=True),square=True, ax=ax, linewidths=.5)
plt.show() 
```

```python
DF_NEW['Log(Sm)']    = np.log10(DF['Stars_Mass'])
DF_NEW['Log(BHM)']   = np.log10(DF['BlackHoleMass'])
DF_NEW['Log(sSFR)']  = np.log10(DF['StarFormationRate']) - np.log10(DF['Stars_Mass'])
DF_NEW['12+Log(O/H)']=  12 + np.log10(DF['SF_Oxygen'])-np.log10(DF['SF_Hydrogen'])-np.log10(16)
DF_NEW['F_g']        = DF['SF_Mass']/(DF['Stars_Mass']+DF['SF_Mass'])
```

```python
DF_NEW.drop(columns = ['SF_Hydrogen','SF_Oxygen', 'SF_Mass', 'NSF_Mass', 'BlackHoleMass',
       'StarFormationRate','Stars_Mass'], inplace = True)
```
