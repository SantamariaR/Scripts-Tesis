```python
# Definimos los límites del bin que nos interesan:

lim_mass   = [10.6,10.8] #bin de masa que queremos
lim_SFmass = 7.75        #cota mínima para el log(SF_Mass)
lim_kappa  = [0.3,0.56]  #limite para valores mínimos y máximos de kappa0.3 0.53

mask_mass  = (table_magni['log(Stars_Mass)']>=lim_mass[0])&(table_magni['log(Stars_Mass)']<=lim_mass[1])
mask_SFmass= (table_magni['log(SF_Mass)']>=lim_SFmass)
```

```python
figure, ax = plt.subplots(figsize=(9,6))

mask_esf  = (table_magni['KappaCoRot']<=lim_kappa[0])&mask_mass&mask_SFmass
mask_disc = (table_magni['KappaCoRot']>=lim_kappa[1])&mask_mass&mask_SFmass
# mask_inter= (table_magni['KappaCoRot']>=lim_kappa[0])&(table_magni['KappaCoRot']<=lim_kappa[1])&mask_mass&mask_SFmass


x_esf = table_magni['log(Stars_Mass)'][mask_esf]
y_esf = 12 + np.log10(table_magni['SF_Oxygen'][mask_esf]/table_magni['SF_Hydrogen'][mask_esf])-np.log10(16)

x_disc = table_magni['log(Stars_Mass)'][mask_disc]
y_disc = 12 + np.log10(table_magni['SF_Oxygen'][mask_disc]/table_magni['SF_Hydrogen'][mask_disc])-np.log10(16)


ax.set_xlabel(r'$log(M_*)$')
ax.set_ylabel(r'$12+log10(O/H)$')

plt.scatter(x_esf,y_esf,color='r',label='esferoidales')
plt.scatter(x_disc,y_disc,color='b',label='discoidales')
plt.grid()
plt.show()
```

```python
# Separamos las poblaciones de galaxias según esferoidales y discoidales

df_temp = table_merger.merge(right=table_magni[['log(Stars_Mass)','KappaCoRot','log(SF_Mass)']],
                             left_on=table_merger['GalaxyID_test'], 
                             right_on=table_magni['GalaxyID']).drop(columns=['key_0']).copy()


mask_mass  = (df_temp['log(Stars_Mass)']>=lim_mass[0])&(df_temp['log(Stars_Mass)']<=lim_mass[1])&(df_temp['log(SF_Mass)']>=lim_SFmass)

mask_esf  = (df_temp['KappaCoRot']<=lim_kappa[0])&mask_mass
mask_disc = (df_temp['KappaCoRot']>=lim_kappa[1])&mask_mass
mask_inter= (df_temp['KappaCoRot']>=lim_kappa[0])&(df_temp['KappaCoRot']<=lim_kappa[1])&mask_mass


table_merger_esf  = df_temp[mask_esf].copy()
table_merger_disc = df_temp[mask_disc].copy()
table_merger_inter= df_temp[mask_inter].copy()

print('Cantidad de galaxias ESFEROIDALES =',len(table_merger_esf))
print('Cantidad de galaxias DISCOIDALES  =',len(table_merger_disc))
```

```python
# lista de redshift permitodos 3, 2, 1, 0.5 y 0

def red_to_snap(z):
    condiciones = [z == 0, z == 0.5, z == 1, z == 2, z == 3]
    resultados  = [    28,       23,     19,     15,     12]
    resultado   = np.select(condiciones, resultados, None)
    return resultado.item()
```

```python
# Intervalo que nos intereza analizar las fusiones y a que level merger:
z = [1,0]  # lista de redshift permitodos 3, 2, 1, 0.5 y 0
Lm= 1.25   #Level of merger que queremos como cota

# Vamos a calcular el porcentaje de galaxias que al menos sufrieron un MM, separando entre esferoidales y discoidales

porc_esf = table_merger_esf.iloc[:,red_to_snap(z[0])+1:red_to_snap(z[1])+2].applymap(lambda x: 1 if x >=Lm  else 0).apply(np.sum, axis = 1).apply(lambda x:1 if x>=1 else 0)

porc_disc= table_merger_disc.iloc[:,red_to_snap(z[0])+1:red_to_snap(z[1])+2].applymap(lambda x: 1 if x >=Lm  else 0).apply(np.sum, axis = 1).apply(lambda x:1 if x>=1 else 0)

print('Porcentaje de galaxias ESFEROIDALES que al menos sufrieron un Lm>='+str(Lm)+' entre z='+str(z[0])+' y z='+str(z[1])+':',np.round(np.sum(porc_esf)/len(porc_esf),2))

print('Porcentaje de galaxias DISCOIDALES que al menos sufrieron un Lm>='+str(Lm)+' entre z='+str(z[0])+' y z='+str(z[1])+' :',np.round(np.sum(porc_disc)/len(porc_disc),2))
```

```python
# Algunos hsitogramas, graficamos la cantidad de mergers seleccionados que sufrieron las galaxias separando según su morfología:
figure, ax = plt.subplots(figsize=(9,6))

fontdict ={'fontsize': 30}

var = table_merger_esf.iloc[:,red_to_snap(z[0])+1:red_to_snap(z[1])+2].applymap(lambda x: 1 if x >=Lm  else 0).apply(np.sum, axis = 1).values

width=1  # Ancho de los bins

bins = np.arange(min(var),max(var)+2,width)

h_var,b = np.histogram(var,bins,density=False)
ax.bar(b[:-1],height=h_var * np.diff(b),width =width,color='pink',edgecolor='purple',capsize=2,label='Cantidad de gal esf:'+str(len(var)))
ax.set_xlabel(r'$Cantidad\ de\ Lm>'+str(Lm)+'$',fontsize=20)
ax.set_ylabel(r'$Eventos$',fontsize=20)
ax.set_xticks(bins)
    
leg=ax.legend(loc='upper right',fontsize=15,handlelength=5)

ax.set_xticklabels(np.round(bins,1),fontsize=10,rotation=0)
plt.title('Eventos de M entre z='+str(z[0])+' y z='+str(z[1])+' para gal esferoidales')
plt.grid()
plt.show()
```
